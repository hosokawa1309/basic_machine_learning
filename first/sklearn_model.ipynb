```
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
#from random import shuffle
import tensorflow as tf
import random
#from numpy import ndarray
import numpy
```

```
N = 300
#0.3の重なりを持つ300個のデータを生成
X , y = datasets.make_moons(N,noise = 0.3)
#X , y = datasets.make_moons(N,noise = 0)
```

```
#実際の値を確認
print(X.shape)
print(X)
print(y.shape)
print(y)
```

```
#Xのうち、y(label) == 0である座標のx座標、y座標を格納
x0, y0 = X[y == 0, 0], X[y == 0,1]
#Xのうち、y(label) == 1である座標のx座標とy座標を格納
x1, y1 = X[y == 1, 0], X[y == 1, 1]

#pltするよ
plt.title("moon")
plt.scatter(x0,y0,label = "label-A")
plt.scatter(x1,y1,label = "label-B")

plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.legend()
```

```
#yを転地している？ 
Y = y.reshape(N, 1)
print(Y.shape)
print(Y)
X_train,  X_test, Y_train, Y_test = train_test_split(X,Y,train_size = 0.8)
```

```
print(X_train)
```

```
#隠れ層の次元数を選択
num_hidden = 2

x = tf.placeholder(tf.float32,shape=[None, 2])
t = tf.placeholder(tf.float32,shape=[None, 1])
```

```
#入力層から隠れ層
W = tf.Variable(tf.truncated_normal([2,num_hidden]))
b = tf.Variable(tf.zeros([num_hidden]))
h = tf.nn.sigmoid(tf.matmul(x,W) + b)

#隠れ層から出力層
V = tf.Variable(tf.truncated_normal([num_hidden,1]))
c = tf.Variable(tf.zeros([1]))
y = tf.nn.sigmoid(tf.matmul(h , V) + c)

cross_entropy = - tf.reduce_sum(t* tf.log(y) + (1 - t) * tf.log(1 - y))
train_step = tf.train.GradientDescentOptimizer(0.05).minimize(cross_entropy)
correct_prediction = tf.equal(tf.cast(tf.greater(y ,0.5), float), t)
```

```
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```
batch_size = 20
n_batches = N
```

```
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
  #デバック用
  X_ = X_train
  Y_ = Y_train
  print(X_[0:5])
  print(Y_[0:5])
  #修正箇所
  numpy.random.shuffle(X_)
  numpy.random.shuffle(Y_)
  #print( X_ == X_train)
  print("ランダム")
  print(X_[0:5])
  print(Y_[0:5])
```

```
for epoch in range(500):
  if epoch%50 == 0:
    print("epoch:",epoch)
  X_ = X_train
  Y_ = Y_train
  
  #修正箇所
  numpy.random.seed(epoch%10)
  numpy.random.shuffle(X_)
  numpy.random.seed(epoch%10)
  numpy.random.shuffle(Y_)
  


  ##XとYを別々の関数でシャッフルする
  for i in range(n_batches):
    #ミニバッチの位置を決定
    start = i * batch_size
    end  = start + batch_size
    
    sess.run(train_step,feed_dict = {
        x: X_[start:end],
        t: Y_[start:end]
    })
```

```
accuracy_rate = accuracy.eval(session = sess,feed_dict={
    x: X_test,
    t: Y_test
})
print('accuracy : ',accuracy_rate)
```
