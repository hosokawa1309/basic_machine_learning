```
import numpy as np
import tensorflow as tf
```

```
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])
```

```
x = tf.placeholder(tf.float32,shape=[None,2])
t = tf.placeholder(tf.float32,shape=[None,1])
```

```
#入力層から隠れ層までも出力
#切断正規分布に従う乱数値をwに与える。
#0で与えてしまうと誤差逆伝搬できない場合がある。
W = tf.Variable(tf.truncated_normal([2,2]))
b = tf.Variable(tf.zeros([2]))
h = tf.nn.sigmoid(tf.matmul(x,W) + b)
```

```
#隠れ層から出力層への出力
V = tf.Variable(tf.truncated_normal([2,2]))
c = tf.Variable(tf.zeros([2]))
y = tf.nn.sigmoid(tf.matmul(h,V) + c)
```

```
#最小化する誤差関数を定義
cross_entropy = - tf.reduce_sum(t * tf.log(y) + (1 - t) * tf.log(1 - y ))
```

```
#学習率0.1で確率的勾配降下法を用いて最小化
train = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
#0.5以上で発火するモデルが正解ラベルtと一致しているか判定
corrent_prediction = tf.equal(tf.cast(tf.greater(y,0.5),float),t)
```

```
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
for epoch in range(4000):
  sess.run(train, feed_dict = {
      x: X,
      t: Y
  })
  if epoch % 1000 == 0:
    print("epoch : ",epoch)
```

```
classified = corrent_prediction.eval(session = sess,feed_dict = {
    x: X,
    t: Y
})

prob = y.eval(session = sess , feed_dict = {
    x: X
})
```

```
print("classified: ")
print(classified)
print()
print("output probability")
print(prob)
```
