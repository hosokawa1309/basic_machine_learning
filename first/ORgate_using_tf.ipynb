```
import numpy as np
import tensorflow as tf
from pprint import pprint
```

```
#変数w、bを定義
w = tf.Variable(tf.zeros([2,1]))
b = tf.Variable(tf.zeros([1]))

#型の確認
pprint(w)
pprint(b)
```

```
#値の入れ物であるx、tを定義
#xには0,1の組み合わせである4通りの入力が格納される
x = tf.placeholder(tf.float32,shape=[None,2])
t = tf.placeholder(tf.float32,shape=[None,1])
y = tf.nn.sigmoid(tf.matmul(x,w) + b)
```

```
#クロスエントロピー誤差関数を定義
cross_entropy = - tf.reduce_sum(t * tf.log(y) + (1 - t) * tf.log(1 - y))
```

```
#誤差関数を最小化するために勾配降下法を行う。学習率は0.1
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
```

```
#予測値yが0.5以上であれば発火が発生し、教師ラベルtと値が一致するか調べる。
corrent_prediction = tf.equal(tf.cast(tf.greater(y,0.5),float),t)
```

```
#入力
X = np.array([[0,0],[0,1],[1,0],[1,1]])
#教師
Y = np.array([[0],[1],[1],[1]])
```

```
#セッションの宣言
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
#学習
for epoch in range(1000):
  sess.run(train_step,feed_dict = {
      x: X,
      t: Y
  })
```

```
#学習結果の確認
classified = corrent_prediction.eval(session = sess,feed_dict={
    x: X,
    t: Y
})
```

```
prob = y.eval(session = sess,feed_dict = {
    x: X
})
```

```
print('classified : ')
pprint(classified)
print()
print('output probability : ')
pprint(prob)
```
